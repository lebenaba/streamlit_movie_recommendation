```python
import pandas as pd

df = pd.read_csv('../data/processed/preprocessed_data.csv')
df_simple = pd.read_csv('../data/processed/preprocessed_data_no_one_hot_encoding.csv')
```

```python
df_simple.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Keeping only the 10 most frequent rows
df_top_10 = df_simple.head(10)

# Plot 1: Bar plot of Ratings
plt.figure(figsize=(10, 6))
plt.bar(df_top_10['title'], df_top_10['rating'])
plt.xlabel('Movie Title')
plt.ylabel('Rating')
plt.title('Ratings of Movies')
plt.show()

# Plot 2: Bar plot of Revenue
plt.figure(figsize=(10, 6))
plt.bar(df_top_10['title'], df_top_10['userId'])
plt.xlabel('Movie Title')
plt.ylabel('Revenue (in millions)')
plt.title('Revenue of Movies')
plt.show()

# Plot 3: Genre Distribution
plt.figure(figsize=(10, 6))
df_top_10['genres'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.ylabel('')
plt.title('Genre Distribution')
plt.show()

# Plot 4: Rating vs Revenue Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df_top_10['rating'], df_top_10['userId'])
plt.xlabel('Rating')
plt.ylabel('Revenue (in millions)')
plt.title('Rating vs Revenue')
plt.show()

# Plot 5: Rating Over Years
plt.figure(figsize=(10, 6))
plt.plot(df_top_10['tags'], df_top_10['rating'], marker='o')
plt.xlabel('Year')
plt.ylabel('Rating')
plt.title('Ratings Over Years')
plt.show()
```
